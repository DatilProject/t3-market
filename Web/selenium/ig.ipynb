```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests

import csv


username = "miathvill9029@gmail.com"
password = "miang9029"
url = "https://www.instagram.com/accounts/login/"
provincias= ["GALAPAGOS"]


def login(username, your_password):
    log_but = browser.find_element_by_class_name("_9GP1n   ")
    time.sleep(1)
    log_but.click()
    browser.implicitly_wait(1)

    usern = browser.find_element_by_name("username")

    usern.send_keys(username)

    passw = browser.find_element_by_name("password")

    passw.send_keys(your_password)

    passw.send_keys(Keys.ENTER)
    time.sleep(1)

def first_picture():
    try:
        pic = browser.find_element_by_class_name("_9AhH0")
        browser.implicitly_wait(1)
        return pic.click()
    except:
         return False

def next_picture():
    try:
        browser.implicitly_wait(1)
        next = browser.find_element_by_xpath("//a[contains(text(),'Next')]")
        return next
    except:
        browser.implicitly_wait(1)
        next = browser.find_element_by_xpath("//a[contains(text(),'Siguiente')]")
        return next

def fecha_2020(fecha):
    fecha2020=fecha[:4]
    if fecha2020=="2020":
        return True
    return False

def quitar_enter(linea):
    texto=""
    palabras=linea.split("\n")
    for pal in palabras:
        texto=texto.strip()+" "+pal.strip()
    return texto

browser=webdriver.Chrome("C:\chromedriver.exe")
browser.get(url)
time.sleep(1)
login(username, password)

#cerrar ventana emergente
time.sleep(1)
windows= browser.find_element_by_xpath("/html/body/div[4]/div/div/div[3]/button[2]")
windows.click()
# browser.maximize_window()
             
    
#Post por perfil
archivo_de_publicaciones_por_provincias_perfil= open("santo_domingo.csv", "a",newline='' ,encoding='utf-8-SIG')
encabezado1=["Provincia","Nombre de usuario","Publicaciones","Seguidores","Seguidos","Descripcion del perfil","Lugar","Fecha","Hora","Descripcion del post"]
salida= csv.writer(archivo_de_publicaciones_por_provincias_perfil,delimiter='|')
salida.writerow(encabezado1)


csvarchivo = open('Publicaciones_por_perfil_Santo_Domingo_De_Los_Tsachilas.csv')  # Abrir archivo csv
entrada = csv.reader(csvarchivo)  # Leer todos los registros
a=0
usu_re = []
for line in entrada:
  time.sleep(2)
#   print(usu_re)
  line="".join(line)
  provincia,usuario,link=line.split("|")
  a=a+1
  if a==1 :
      continue

  if usuario in usu_re:
      continue

  usu_re.append(usuario)
  print(usuario)
  time.sleep(1)
  browser.get(link)
  if first_picture() == False:
      continue
  try:
      browser.implicitly_wait(3)
      browser.find_element_by_xpath("/html/body/div/div[1]/div/div/h2").text # =="Sorry, this page isn't available."
  except:
      browser.implicitly_wait(1)
      descripcion_perfil=browser.find_element_by_class_name("-vDIg").text
      descripcion_perfil=quitar_enter(descripcion_perfil)
      descripcion_perfil=descripcion_perfil.replace("|", "")
      descripcion_perfil=descripcion_perfil.replace(",", "")
      descripcion_perfil=descripcion_perfil.replace('"', "")
      usu=link[26:-1]
      browser.implicitly_wait(1)
      d_publicados=browser.find_elements_by_class_name("-nal3 ")
      lista = []
      for elem in d_publicados:
          browser.implicitly_wait(1)
          elem = elem.find_element_by_class_name("g47SY ").text
          lista.append(elem)
#       print(lista)
      if lista[0]=="" and lista[1]==""  and lista[2]=="":
          browser.implicitly_wait(1)
          r=requests.get(link)
          meta=BeautifulSoup(r.text, "html.parser")
          meta = meta.find("meta", property="og:description")
          d_publicados=meta.attrs['content']
          d_publicados=d_publicados.split("-")
          d_publicad=d_publicados[0].strip().split(" ")
          try:
              if d_publicad[4].isalpha() or  d_publicad[0].isalpha() or d_publicad[2].isalpha():
                  lista=["","",""]
                  lista = "|".join(lista)
              else:
                  lista="{} posts|{} followers|{} following".format(d_publicad[4],d_publicad[0],d_publicad[2])                              
          except:
                lista=["","",""]
                lista = "|".join(lista)

      else:
          lista[0] = lista[0] + " " + "posts"
          lista[1] = lista[1] + " " + "followers"
          lista[2] = lista[2] + " " + "following"
          lista = "|".join(lista)
      lista = lista.replace(",", ".")
      provincia=provincia.replace("Ã‘","Ñ")
      datos= provincia+'|'+usu+'|'+lista+'|'+descripcion_perfil
      time.sleep(1)
      i=0
      x=0
      while True:
            try:
              i=i+1
              browser.implicitly_wait(1)
              time.sleep(2)
              name = browser.find_element_by_class_name("e1e1d")
              browser.implicitly_wait(1)
              nombre_usuario = name.text
              nombre_usuario = nombre_usuario.strip("Verified\n")
              browser.implicitly_wait(1)
              ubicacion = browser.find_element_by_class_name("JF9hh")
              browser.implicitly_wait(1)
              if ubicacion.text != "":
                  browser.implicitly_wait(1)
                  ubicacion = browser.find_element_by_class_name("JF9hh").text
                  browser.implicitly_wait(1)
                  ubicacion=ubicacion.replace(",", "")
              else:
                  ubicacion = ""
              browser.implicitly_wait(1)
              descripcion = browser.find_element_by_class_name("EtaWk").text

              if descripcion != "":
                  browser.implicitly_wait(1)
                  descripcion = browser.find_element_by_class_name("C4VMK")
                  descripcion = descripcion.text
                  if nombre_usuario in descripcion:
                      browser.implicitly_wait(1)
                      descripcion = descripcion.strip(nombre_usuario + "Verified\n")
                      try:
                          browser.implicitly_wait(1)
                          tiempo = browser.find_element_by_xpath("/html/body/div[4]/div[2]/div/article/div[2]/div[1]/ul/div/li/div/div/div[2]/div/div/time").text
                          descripcion = descripcion.strip(tiempo + "\n")
                          descripcion = quitar_enter(descripcion)
                          descripcion = descripcion.replace("|", "")
                          descripcion = descripcion.replace(",", "")
                          descripcion = descripcion.replace('"', "")

                      except :
                          browser.implicitly_wait(1)
                          tiempo = browser.find_element_by_xpath("/html/body/div[4]/div[2]/div/article/div[2]/div[1]/ul/ul/div/li/div/div[1]/div[2]/div/div/a/time").text
                          descripcion = descripcion.strip(tiempo + "\n")
                          descripcion = quitar_enter(descripcion)
                          descripcion = descripcion.replace("|", "")
                          descripcion = descripcion.replace(",", "")
                          descripcion = descripcion.replace('"', "")


                  else:
                      descripcion = ""
              else:
                  descripcion = ""

              browser.implicitly_wait(1)
              datetime = browser.find_element_by_class_name("c-Yi7").find_element_by_tag_name("time").get_attribute("datetime")
              datetime = datetime.split("T")
              fecha = datetime[0]
              hora = datetime[1][:8]
              if fecha_2020(fecha) == False :
                  break
              else:
                  datos_usuario_provincia = datos + '|' + ubicacion + '|' + fecha + '|' + hora + '|' + descripcion +  "\n"
                  archivo_de_publicaciones_por_provincias_perfil.write(datos_usuario_provincia)
    #               print(datos_usuario_provincia)
                  try:
                      time.sleep(2)
                      next_picture().click()
                  except:
                      break
            except:
                time.sleep(1)
                try:
                    time.sleep(2)
                    next_picture().click()
                except:
                    x+=1  
                    if x>3:
                        break
                    continue
                
                
browser.close()
archivo_de_publicaciones_por_provincias_perfil.close()
csvarchivo.close()
```
