```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('data/santa_elena.csv',sep='|')
df.head()
```

```python
lista_casa=["casa","casita","cuarentena","quedate"]

def contar_palabras_de_quedarte_en_casa(post):
    i = 0
    for palabra in lista_casa:
        cantidad= str(post).count(palabra)
        i += cantidad
    return i


df["Palabras cantidad"]=df['Descripcion del post'].apply(contar_palabras_de_quedarte_en_casa)
```

```python
df[df["Palabras cantidad"]!=0]
```

```python
from nltk.corpus import stopwords

texts=df['Descripcion del post']
texts = texts[texts.notnull()]
l = []

for frase in texts:
    l.extend(frase.split())

sws = stopwords.words('spanish')

words = list(filter(lambda word: word not in sws, l))
```

```python
cv = CountVectorizer()
cv_fit=cv.fit_transform(words)

voc = cv.vocabulary_
```

```python
a = pd.DataFrame(voc.keys())
a['b'] = voc.values()
a.sort_values(by='b', ascending=False, inplace=True)
a.head()
```
